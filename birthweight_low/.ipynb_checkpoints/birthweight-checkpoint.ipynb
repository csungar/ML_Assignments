```python
import pandas as pd # data science essentials
import matplotlib.pyplot as plt # essential graphical output
import seaborn as sns # enhanced graphical output
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf


# setting pandas print options
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)


# specifying file name
file = './birthweight.xlsx'


# reading the file into Python
bw = pd.read_excel(file)


# outputting the first ten rows of the dataset
bw.head(n=10)



```

```python
bw.info()
```

```python
sns.histplot(data  = bw,
             x     = "bwght",
             kde   = True)


# title and axis labels
plt.title(label   = "Dist of birthweight")
plt.xlabel(xlabel = "birthweight") # avoiding using dataset labels
plt.ylabel(ylabel = "Count")


# displaying the histogram
plt.show()
```

```python
lm_best = smf.ols(formula =  """ bwght ~ npvis + 
                                         fage +
                                         fage +
                                         npvis +
                                         omaps +
                                         fmaps +
                                         male +
                                         cigs +
                                         male +
                                         fwhte +
                                         fblck
                                             """,
                                data = bw)


# Step 2: FIT the data into the model object
results = lm_best.fit()


# Step 3: analyze the SUMMARY output
print(results.summary())
```

```python
for col in bw:
    if bw[col].isnull().sum() > 0:
         bw['missing_'+ col] = bw[col].isnull().astype(int)
            
bw['missing_values_sum'] = 0

for col in bw:
    if col[:2] == 'missing_':
        bw['missing_values_sum'] = bw[col] + bw['missing_values_sum']

print(f"""
Sums of Missing Value Flags
--------------------------
{bw.iloc[ : , -10:-1 ].sum(axis = 0)}

""")
```

```python
for col in bw: 
    if bw[col].isnull().sum() > 0 and bw[col].dtypes == 'int64':
        median = bw[col].median()
        mean = bw[col].mean()
        if median < mean:
            bw[col].fillna(value = median,
                              inplace = True)
        else:
            bw[col].fillna(values = mean,
                               inplace = True)

#If it is an object will be repalce with the mode
    elif bw[col].isnull().sum() > 0 and bw[col].dtypes =='float64':
        mode = bw[col].mode()
        bw[col].fillna(value = mode[0],
                           inplace = True)

print(bw.isnull().any())
```
