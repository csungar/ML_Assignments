```python
#Replacing the MAGE to the Categorical Data with Bins(<26, 26-28, 29-32, >=33)
if 'mage_bins' not in df.columns:
    mage_bins = []  #creating helping column
    
    for i in df['mage']:
        if i < 40:
            mage_bins.append("less40")
        else:
            mage_bins.append("40plus")
    df.insert(loc = 17, column = 'mage_bins', value = mage_bins)  #inserting helping column to the dataset
    


#Replacing the MEDUC to the Categorical Data with Bins(<12, 12, 13-15, >=16)
if 'meduc_bins' not in df.columns:
    meduc_bins = []  #creating helping column

    for i in df['meduc']:
        if i < 12:
            meduc_bins.append("less12")
        elif i >= 12 and 16 > i:
            meduc_bins.append("12to15")
        else:
            meduc_bins.append("16plus")
    df.insert(loc = 18, column = 'meduc_bins', value = meduc_bins)  #inserting helping column to the dataset
    
#Replacing the MONPRE to the Categorical Data with Bins(0-1, 2, >=3)
if 'monpre_bins' not in df.columns:
    monpre_bins = []  #creating helping column

    for i in df['monpre']:
        if i < 4 :
            monpre_bins.append("less4")
        else:
            monpre_bins.append("4plus")
    df.insert(loc = 19, column = 'monpre_bins', value = monpre_bins)  #inserting helping column to the dataset
    
#Replacing the NPVIS to the Categorical Data with Bins(<10, 10-11, 12, >=13)
if 'npvis_bins' not in df.columns:
    npvis_bins = []  #creating helping column

# it is assumed that a healthy pregnancy involves 9 prenatal visits (once per month)
# if this number is more or less then the pregnancy is odd
    for i in df['npvis']:
        if i < 8:
            npvis_bins.append("less8")
        elif i < 13 and i >= 8:
            npvis_bins.append("normal")
        else:
            npvis_bins.append("13plus")

    df.insert(loc = 20, column = 'npvis_bins', value = npvis_bins)  #inserting helping column to the dataset
    
   #Replacing the FEDUC to the Categorical Data with Bins(<12, 12-13, 14-15, >=16)
if 'feduc_bins' not in df.columns:
    feduc_bins = []  #creating helping column

    for i in df['feduc']:
        if i < 12:
            feduc_bins.append("less12")
        elif i >= 12 and i > 14:
            feduc_bins.append("12to13")
        elif i >= 14 and 16 > i:
            feduc_bins.append("14to15")
        else:
            feduc_bins.append("16plus")
    df.insert(loc = 22, column = 'feduc_bins', value = feduc_bins)  #inserting helping column to the dataset
    
    
#Replacing the CIGS to the Categorical Data with Bins(0, not0)
if 'cigs_bins' not in df.columns:
    cigs_bins = []  #creating helping column

    for i in df['cigs']:
        if i == 0:
            cigs_bins.append("non_smoker")
        else:
            cigs_bins.append("smoker")
    df.insert(loc = 25, column = 'cigs_bins', value = cigs_bins)  #inserting helping column to the dataset
    
#Replacing the DRINKS to the Categorical Data with Bins(0, not0)
if 'drink_bins' not in df.columns:
    drink_bins = []  #creating helping column

    for i in df['drink']:
        if i == 0:
            drink_bins.append("non_drinker")
        else:
            drink_bins.append("drinker")
    df.insert(loc = 26, column = 'drink_bins', value = drink_bins)  #inserting helping column to the dataset
    
df.head(n = 10)
```
